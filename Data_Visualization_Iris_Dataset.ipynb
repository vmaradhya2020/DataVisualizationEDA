```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset
url = '/content/sample_data/Iris.csv'
data = pd.read_csv(url)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# Drop the 'Id' column as it's not relevant for visualization
data = data.drop(columns=['Id'])
```

```python
# Set a style for the plots
sns.set(style="whitegrid")
```

```python
# 1. Distribution of Sepal Lengths and Widths Among Species
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="SepalLengthCm", hue="Species", kde=True, element="step", bins=15)
plt.title("Distribution of Sepal Lengths Among Iris Species")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="SepalWidthCm", hue="Species", kde=True, element="step", bins=15)
plt.title("Distribution of Sepal Widths Among Iris Species")
plt.show()
```

```python
# 2. Variation of Petal Length Across Species
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x="Species", y="PetalLengthCm")
plt.title("Variation of Petal Length Across Iris Species")
plt.show()
```

```python
# 3. Relationship Between Sepal Length and Petal Length
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x="SepalLengthCm", y="PetalLengthCm", hue="Species")
plt.title("Relationship Between Sepal Length and Petal Length")
plt.show()
```

```python
# 4. Variation of Petal Width Across Species
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x="Species", y="PetalWidthCm")
plt.title("Variation of Petal Width Across Iris Species")
plt.show()
```

```python
# 5. Pairwise Relationships Between Features
sns.pairplot(data, hue="Species", diag_kind="kde", markers=["o", "s", "D"])
plt.suptitle("Pairwise Relationships Between Features", y=1.02)
plt.show()
```

```python
# 6. Distinguishability of Species Based on Measurements
plt.figure(figsize=(12, 6))
sns.heatmap(data.drop(columns=['Species']).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Features")
plt.show()
```
