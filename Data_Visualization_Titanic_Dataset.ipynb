# **Titanic Dataset Visualization Notebook**

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
data = pd.read_csv(url)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
### Question 1: What is the overall survival rate?
# Plotting the survival count
sns.countplot(x='Survived', data=data, palette='viridis')
plt.title('Survival Count (0 = Not Survived, 1 = Survived)')
plt.show()

# Calculate survival rate
survival_rate = data['Survived'].mean() * 100
print(f"Overall Survival Rate: {survival_rate:.2f}%")
```

```python
### Inference:
# - Around survival_rate: 38.38% of passengers survived.
```

```python
### Question 2: How does gender impact survival rate?
# Plotting survival by gender
sns.countplot(x='Survived', hue='Sex', data=data, palette='viridis')
plt.title('Survival by Gender')
plt.show()

# Calculate survival rate by gender
survival_gender = data.groupby('Sex')['Survived'].mean() * 100
print("Survival Rate by Gender:\n", survival_gender)
```

```python
### Inference:
# - Females had a significantly higher survival rate compared to males.
```

```python
### Question 3: What is the age distribution of passengers?
# Plotting age distribution
sns.histplot(data['Age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
### Inference:
# - Most passengers were aged between 20 and 40.
```

```python
### Question 4: How does survival rate vary by age?
# Plotting age vs survival
sns.boxplot(x='Survived', y='Age', data=data, palette='coolwarm')
plt.title('Age vs Survival')
plt.show()
```

```python
### Inference:
# - Younger passengers had higher survival rates.
```

```python
### Question 5: What is the survival rate by passenger class?
# Plotting survival by passenger class
sns.countplot(x='Pclass', hue='Survived', data=data, palette='pastel')
plt.title('Survival by Passenger Class')
plt.show()

# Calculate survival rate by class
survival_class = data.groupby('Pclass')['Survived'].mean() * 100
print("Survival Rate by Passenger Class:\n", survival_class)
```

```python
### Inference:
# - Passengers in 1st class had the highest survival rate, while those in 3rd class had the lowest.
```

```python
### Question 6: How does fare vary among passenger classes?
# Plotting fare distribution by class
sns.boxplot(x='Pclass', y='Fare', data=data, palette='muted')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
### Inference:
# - Passengers in 1st class paid significantly higher fares compared to other classes.
```

```python
### Question 7: How does embarkation port affect survival?
# Plotting survival by embarkation port
sns.countplot(x='Embarked', hue='Survived', data=data, palette='bright')
plt.title('Survival by Embarkation Port')
plt.show()
# Calculate survival rate by embarkation port
survival_embarked = data.groupby('Embarked')['Survived'].mean() * 100
print("Survival Rate by Embarkation Port:\n", survival_embarked)
```

```python
### Inference:
# - Passengers embarking from port 'C' had the highest survival rate.
```

```python
### Question 8: Does having siblings/spouses aboard impact survival?
# Plotting survival by siblings/spouses aboard
sns.countplot(x='SibSp', hue='Survived', data=data, palette='Set2')
plt.title('Survival by Siblings/Spouses Aboard')
plt.show()
```

```python
### Inference:
# - Passengers with 1 siblings/spouses aboard had higher survival rates.
```

```python
### Question 9: Does having parents/children aboard impact survival?
# Plotting survival by parents/children aboard
sns.countplot(x='Parch', hue='Survived', data=data, palette='Set1')
plt.title('Survival by Parents/Children Aboard')
plt.show()
```

```python
### Inference:
# - Passengers with 1 and 3 parents/children aboard had higher survival rates.
```

```python
### Question 10: What is the relationship between fare and survival?
# Plotting fare vs survival
sns.boxplot(x='Survived', y='Fare', data=data, palette='cool')
plt.title('Fare vs Survival')
plt.show()
```

```python
### Inference:
# - Passengers who survived tended to have paid higher fares.
```

```python
### Question 11: How does cabin information correlate with survival?
# Extracting cabin presence
sns.countplot(x=data['Cabin'].notnull(), hue=data['Survived'], palette='husl')
plt.title('Survival Based on Cabin Information')
plt.xlabel('Cabin Known (True/False)')
plt.show()
```

```python
### Inference:
# - Passengers with known cabin information had higher survival rates.
```

```python
### Question 12: Is there a relationship between ticket class and gender?
# Plotting ticket class by gender
sns.countplot(x='Pclass', hue='Sex', data=data, palette='viridis')
plt.title('Ticket Class Distribution by Gender')
plt.show()
```

```python
### Inference:
# - More males traveled in 3rd class, while females were more evenly distributed.
```
